```python
def mape(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    mape_arr=(np.abs(actual - pred) / actual)*100
    return (mape_arr,np.mean( (np.abs(actual - pred) / actual) * 100))
```
